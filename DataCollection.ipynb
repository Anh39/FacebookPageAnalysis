```python
import facebook_scraper as fbs
import numpy as np
import pandas as pd
import time
import random
```

```python
FANPAGE_LINK = 'unity3d'
FOLDER_PATH = 'Data/'
COOOKIE_PATH = 'cookies.txt'

PAGES_NUMBER = 500
TIME_OUT = 30
```

```python
post_list = []
posts = fbs.get_posts(FANPAGE_LINK,
                      options={"comments": True, "reactions": True, "allow_extra_requests": True,"progress":False},
                      extra_info = True, pages = PAGES_NUMBER,cookies = COOOKIE_PATH,timeout = TIME_OUT)
```

```python
i = 0
for post in posts:
    post_list.append(post)
    print(len(post_list))
    if (len(post_list) >= PAGES_NUMBER):
        break
    time.sleep(random.randrange(10,30))
    i+=1
    if (i == 100):
        time.sleep(25 * 3600)
```

```python
print(len(post_list))
```

```python
post_df_full = pd.DataFrame(columns=post_list[0].keys(), index=range(len(post_list)), data=post_list)
```

```python
post_df_full.info()
post_df_full.describe()
```

```python
fieldToExtract = ['post_id','post_text','time','reaction_count','reactions','shares','comments','comments_full']
post_df = post_df_full[fieldToExtract]
```

```python
post_df.info()
post_df.describe()
```

```python
path=FOLDER_PATH + FANPAGE_LINK + "Extracted.csv"
post_df.to_csv(path, index=False)
path=FOLDER_PATH + FANPAGE_LINK + "Raw.csv"
post_df_full.to_csv(path, index=False)
```
